```python
import torch, json
dev_path='/media/melissa/EXTERNAL_USB/docred_output/dev.pt'
test_path='/media/melissa/EXTERNAL_USB/docred_output/test.pt'
train_path='/media/melissa/EXTERNAL_USB/docred_output/train.pt'
pre_train_path='/media/melissa/EXTERNAL_USB/docred_output/pretrain.pt'
rel2id_path ='/media/melissa/EXTERNAL_USB/DocRED/DocRED_baseline_metadata/rel2id.json'
```

```python
with open(rel2id_path) as f:
    rel2id = json.load(f)
```

```python
na_id = rel2id['Na']
```

```python
dev = torch.load(dev_path)
test = torch.load(test_path)
train = torch.load(train_path)
pretrain = torch.load(pre_train_path)
```

```python
def neg_pos_ratio(data):
    pos_triples = 0
    neg_triples = 0

    for ex in data:
        for _, _, r in ex['labels']:
            if r == na_id:
                neg_triples += 1
            else:
                pos_triples += 1

    print("TRIPLE-LEVEL COUNTS")
    print(f"  Positive relation triples: {pos_triples}")
    print(f"  NA (negative) triples     : {neg_triples}")
    print(f"  Pos:Neg ratio             : {pos_triples / max(1, neg_triples):.4f}")
```

```python
neg_pos_ratio(train)
```

```python
neg_pos_ratio(test)
```

```python
neg_pos_ratio(pretrain)
```

```python
from collections import Counter
rel_counter = Counter()
for ex in train:
    for _, _, r in ex['labels']:
        rel_counter[r] += 1

# Map ids back to names
id2rel = {v:k for k,v in rel2id.items()}

print("\nTOP RELATIONS:")
for r_id, count in rel_counter.most_common():
    print(f"{id2rel[r_id]:<20} {count}")
```

```python
from collections import Counter
label_counter = Counter()

for doc in train:
    for h, t, r in doc["labels"]:
        label_counter[r] += 1

print("Top 10 most frequent relations:")
print(label_counter.most_common(10))
```
